```python
df["Price_INR"] = df["Price in INR"].apply(extract_price)
```

```python
#  Distribution of mobile prices
plt.figure(figsize=(10,6))
sns.histplot(df["Price_INR"], bins=50, kde=True, color="skyblue")
plt.title("Distribution of Mobile Prices", fontsize=16)
plt.xlabel("Price (INR)", fontsize=12)
plt.ylabel("Number of Mobiles", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
# Define price bins
bins = [0, 5000, 10000, 15000, 20000, 30000, 50000, 100000, df["Price_INR"].max()]
labels = ["0-5k", "5k-10k", "10k-15k", "15k-20k", "20k-30k", "30k-50k", "50k-100k", "100k+"]

# Create new column for price ranges
df["Price_Range"] = pd.cut(df["Price_INR"], bins=bins, labels=labels, include_lowest=True)

# Count number of mobiles per price range
price_counts = df["Price_Range"].value_counts().sort_index()

# Plot bar chart
plt.figure(figsize=(10,6))
sns.barplot(x=price_counts.index, y=price_counts.values, palette="Blues_d")
plt.title("Number of Mobiles in Each Price Range", fontsize=16)
plt.xlabel("Price Range (INR)", fontsize=12)
plt.ylabel("Number of Mobiles", fontsize=12)
plt.show()
```

```python
# Convert RAM column into numeric (GB)
def extract_numeric(value):
    """Extract numeric value from strings like '8 GB' """
    if pd.isnull(value):
        return None
    import re
    match = re.search(r"\d+", str(value).replace(",", ""))
    return int(match.group()) if match else None

df["RAM_GB"] = df["RAM"].apply(extract_numeric)

```

```python
#  Distribution of RAM across all devices (Boxplot)

plt.figure(figsize=(8,5))
sns.boxplot(x=df["RAM_GB"], color="skyblue")
plt.title("Distribution of RAM Across All Devices", fontsize=16)
plt.xlabel("RAM_(GB)", fontsize=12)
plt.show()
```

```python
df["Battery_mAh"] = df["Battery"].apply(extract_numeric)

#  Distribution of Battery Capacity

plt.figure(figsize=(10,6))
sns.histplot(df["Battery_mAh"], bins=40, kde=True, color="green")
plt.title("Distribution of Battery Capacity", fontsize=16)
plt.xlabel("Battery Capacity (mAh)", fontsize=12)
plt.ylabel("Number of Mobiles", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
#  How does price vary with RAM? (Scatterplot)

plt.figure(figsize=(10,6))
sns.scatterplot(x=df["RAM_GB"], y=df["Price_INR"], alpha=0.6, color="blue", edgecolor="w")

plt.title("Price vs RAM", fontsize=16)
plt.xlabel("RAM (GB)", fontsize=12)
plt.ylabel("Price (INR)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
# Create a new column "Supports_5G" by checking if '5G' appears in Processor or Phone Name
df["Supports_5G"] = df["Processor"].str.contains("5G", case=False, na=False) | df["Phone Name"].str.contains("5G", case=False, na=False)

# Group by 5G support and calculate average price
avg_price_5g = df.groupby("Supports_5G")["Price_INR"].mean()

print(" Average Price Comparison (INR):")
print(avg_price_5g)

# Plot bar chart
plt.figure(figsize=(7,5))
sns.barplot(x=avg_price_5g.index.map({True:"5G Phones", False:"Non-5G Phones"}),
            y=avg_price_5g.values, palette="Set2")

plt.title("Average Price: 5G vs Non-5G Phones", fontsize=16)
plt.ylabel("Average Price (INR)", fontsize=12)
plt.xlabel("")
plt.show()
```

```python
# Extract brand from Phone Name (first word before space)
df["Brand"] = df["Phone Name"].str.split().str[0]

# Group by brand and calculate average price
avg_price_brand = df.groupby("Brand")["Price_INR"].mean().sort_values(ascending=False)

# Show top 10 brands by average price
print(" Average Price of Top 10 Brands:\n")
print(avg_price_brand.head(10))

# Plot bar chart
plt.figure(figsize=(12,6))
sns.barplot(x=avg_price_brand.head(10).index, y=avg_price_brand.head(10).values, palette="coolwarm")
plt.title("Average Price of Phones by Brand (Top 10)", fontsize=16)
plt.xlabel("Brand", fontsize=12)
plt.ylabel("Average Price (INR)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
df["RAM_GB"] = df["RAM"].apply(extract_numeric)
df["ROM_GB"] = df["ROM/Storage"].apply(extract_numeric)

# Group by RAM and Storage, calculate average price
ram_rom_price = df.groupby(["RAM_GB", "ROM_GB"])["Price_INR"].mean().reset_index()

# Pivot the table correctly
pivot_table = ram_rom_price.pivot(index="RAM_GB", columns="ROM_GB", values="Price_INR")

# Plot the heatmap
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Average Price by RAM and Storage", fontsize=16)
plt.xlabel("Storage (GB)", fontsize=12)
plt.ylabel("RAM (GB)", fontsize=12)
plt.show()
```

```python
# Select only numeric columns for correlation
corr_features = df[["RAM_GB", "ROM_GB", "Battery_mAh", "Price_INR"]]

# Compute correlation matrix
corr_matrix = corr_features.corr()

# Plot correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap: RAM, Storage, Battery vs Price", fontsize=16)
plt.show()
```

```python
# Extract brand from Phone Name (first word)
df["Brand"] = df["Phone Name"].str.split().str[0]

# Filter premium phones (Price > 50k)
premium_phones = df[df["Price_INR"] > 50000]

# Count number of models per brand in premium segment
premium_brands = premium_phones["Brand"].value_counts().head(5)

# Print result
print("📊 Top 5 Brands in Premium (₹50k+) Segment:\n")
print(premium_brands)

# Plot bar chart
plt.figure(figsize=(8,5))
sns.barplot(x=premium_brands.index, y=premium_brands.values, palette="magma")
plt.title("Top 5 Brands in Premium (₹50k+) Market", fontsize=16)
plt.xlabel("Brand", fontsize=12)
plt.ylabel("Number of Models", fontsize=12)
plt.show()
```

```python
# Define price segments
def price_segment(price):
    if price <= 15000:
        return "Low-end (≤15k)"
    elif price <= 40000:
        return "Mid-range (15k-40k)"
    else:
        return "Premium (>40k)"

# Apply function to create new column
df["Price_Segment"] = df["Price_INR"].apply(price_segment)

# Count number of models per segment
segment_counts = df["Price_Segment"].value_counts()

print("📊 Number of Models in Each Price Segment:\n")
print(segment_counts)

# Plot bar chart
plt.figure(figsize=(8,5))
sns.barplot(x=segment_counts.index, y=segment_counts.values, palette="viridis")
plt.title("Number of Mobile Models in Each Price Segment", fontsize=16)
plt.xlabel("Price Segment", fontsize=12)
plt.ylabel("Number of Models", fontsize=12)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
# Select features
features = df[["RAM_GB", "ROM_GB", "Price_INR"]].dropna()
# Scale the data (important for clustering)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)  # you can adjust clusters
features["Cluster"] = kmeans.fit_predict(scaled_features)

# Scatterplot: RAM vs Price, color by cluster, bubble size = Storage
plt.figure(figsize=(10,6))
scatter = plt.scatter(
    features["RAM_GB"], 
    features["Price_INR"], 
    c=features["Cluster"], 
    s=features["ROM_GB"], 
    alpha=0.6
)

plt.title("Clusters of Mobiles (RAM vs Price, Bubble = Storage)")
plt.xlabel("RAM (GB)")
plt.ylabel("Price (INR)")
plt.colorbar(scatter, label="Cluster")
plt.show()
```
